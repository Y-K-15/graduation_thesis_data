 町丁目データの緯度経度を、価格データにマージする。
- keyは、町丁目の名前
  - 町丁目データ(`./kyoto_city_chochomoku.csv`)においてはS_NAME列
  - 価格データ(`../price/combined_price_data.csv`)においてはDistrictName列
- 価格データに、町丁目データのX_CODE列とY_CODE列をマージする
  - 以下のように列名を変更
    - X_CODE→longitude
    - Y_CODE→latitude

```python
import pandas as pd

 ファイルパスの設定
chochomoku_csv_path = './kyoto_city_chochomoku.csv'
price_csv_path = '../price/combined_price_data.csv'
output_csv_path = './merged_price_data.csv'
```

```python
 1. 町丁目データの読み込み
try:
    chochomoku_df = pd.read_csv(chochomoku_csv_path, encoding='utf-8-sig')
    print(f"町丁目データ '{chochomoku_csv_path}' の読み込みに成功しました。")
except FileNotFoundError:
    print(f"エラー: ファイル '{chochomoku_csv_path}' が見つかりません。")
    exit(1)
except Exception as e:
    print(f"町丁目データの読み込み中にエラーが発生しました: {e}")
    exit(1)

 必要な列が存在するか確認
required_chochomoku_columns = ['S_NAME', 'X_CODE', 'Y_CODE']
missing_columns = [col for col in required_chochomoku_columns if col not in chochomoku_df.columns]
if missing_columns:
    print(f"エラー: 町丁目データに以下の必要な列が存在しません: {missing_columns}")
    exit(1)

 2. 価格データの読み込み
try:
    price_df = pd.read_csv(price_csv_path, encoding='utf-8-sig')
    print(f"価格データ '{price_csv_path}' の読み込みに成功しました。")
except FileNotFoundError:
    print(f"エラー: ファイル '{price_csv_path}' が見つかりません。")
    exit(1)
except Exception as e:
    print(f"価格データの読み込み中にエラーが発生しました: {e}")
    exit(1)

 必要な列が存在するか確認
required_price_columns = ['DistrictName']
missing_price_columns = [col for col in required_price_columns if col not in price_df.columns]
if missing_price_columns:
    print(f"エラー: 価格データに以下の必要な列が存在しません: {missing_price_columns}")
    exit(1)

 3. 価格データから「学区」を含むDistrictNameの行を削除
initial_price_count = len(price_df)
price_df = price_df[~price_df['DistrictName'].str.contains('学区', na=False)]
filtered_price_count = len(price_df)
removed_count = initial_price_count - filtered_price_count
print(f"'DistrictName' に '学区' を含む行を {removed_count} 件削除しました。")
 'DistrictName' に '学区' を含む行を 16294 件削除しました。

 4. 町丁目データの列名を変更
chochomoku_df_renamed = chochomoku_df.rename(columns={
    'X_CODE': 'longitude',
    'Y_CODE': 'latitude'
})

 必要な列のみを抽出
chochomoku_subset = chochomoku_df_renamed[['S_NAME', 'longitude', 'latitude']]

```

```python
 5. 部分一致によるマージ

def find_coordinates(district_name, chochomoku_df):
    """
    DistrictNameがS_NAMEに含まれるかをチェックし、対応する緯度経度を返す関数
    """
    if not isinstance(district_name, str):
        print(f"警告: DistrictName '{district_name}' が文字列ではありません。スキップします。")
        return pd.Series({'longitude': None, 'latitude': None})
    
     S_NAMEにDistrictNameが含まれる行をフィルタリング（部分一致）
    match = chochomoku_df[chochomoku_df['S_NAME'].str.contains(district_name, na=False)]
    
    if len(match) == 1:
         一意にマッチした場合
        print("一意にマッチしました。")
        return pd.Series({'longitude': match['longitude'].values[0], 'latitude': match['latitude'].values[0]})
    elif len(match) > 1:
         複数マッチした場合は最初のマッチを使用（必要に応じて他の処理を追加）
        print(f"警告: DistrictName '{district_name}' が複数のS_NAMEにマッチしました。最初のマッチを使用します。")
        return pd.Series({'longitude': match['longitude'].values[0], 'latitude': match['latitude'].values[0]})
    else:
         マッチしない場合
        print(f"警告: DistrictName '{district_name}' に対応するS_NAMEが見つかりませんでした。")
        return pd.Series({'longitude': None, 'latitude': None})

 価格データのDistrictNameを文字列に変換
price_df['DistrictName'] = price_df['DistrictName'].astype(str)

 町丁目データのS_NAMEも必要に応じて文字列に変換
chochomoku_subset['S_NAME'] = chochomoku_subset['S_NAME'].astype(str)

 価格データに対して部分一致で緯度経度を取得
price_df[['longitude', 'latitude']] = price_df['DistrictName'].apply(lambda x: find_coordinates(x, chochomoku_subset))

 6. マージ後の確認
missing_coords = price_df['longitude'].isnull() | price_df['latitude'].isnull()
if missing_coords.any():
    missing_districts = price_df.loc[missing_coords, 'DistrictName'].unique()
    print(f"警告: 以下のDistrictNameに対応する町丁目データが見つかりませんでした: {missing_districts}")

 7. マージ結果の保存
try:
    price_df.to_csv(output_csv_path, index=False, encoding='utf-8-sig')
    print(f"マージされたデータを '{output_csv_path}' に正常に出力しました。")
except Exception as e:
    print(f"CSVへの出力中にエラーが発生しました: {e}")
    exit(1)
```

```python
import pandas as pd

 ファイルパスの設定
chochomoku_csv_path = './kyoto_city_chochomoku.csv'
price_csv_path = '../price/combined_price_data.csv'
output_csv_path = './merged_price_data_type.csv'

 1. 町丁目データの読み込み
try:
    chochomoku_df = pd.read_csv(chochomoku_csv_path, encoding='utf-8-sig')
    print(f"町丁目データ '{chochomoku_csv_path}' の読み込みに成功しました。")
except FileNotFoundError:
    print(f"エラー: ファイル '{chochomoku_csv_path}' が見つかりません。")
    exit(1)
except Exception as e:
    print(f"町丁目データの読み込み中にエラーが発生しました: {e}")
    exit(1)

 必要な列が存在するか確認
required_chochomoku_columns = ['S_NAME', 'X_CODE', 'Y_CODE']
missing_columns = [col for col in required_chochomoku_columns if col not in chochomoku_df.columns]
if missing_columns:
    print(f"エラー: 町丁目データに以下の必要な列が存在しません: {missing_columns}")
    exit(1)

 2. 価格データの読み込み
try:
    price_df = pd.read_csv(price_csv_path, encoding='utf-8-sig')
    print(f"価格データ '{price_csv_path}' の読み込みに成功しました。")
except FileNotFoundError:
    print(f"エラー: ファイル '{price_csv_path}' が見つかりません。")
    exit(1)
except Exception as e:
    print(f"価格データの読み込み中にエラーが発生しました: {e}")
    exit(1)

 必要な列が存在するか確認
required_price_columns = ['DistrictName']
missing_price_columns = [col for col in required_price_columns if col not in price_df.columns]
if missing_price_columns:
    print(f"エラー: 価格データに以下の必要な列が存在しません: {missing_price_columns}")
    exit(1)

 3. 価格データから「学区」を含むDistrictNameの行を削除
initial_price_count = len(price_df)
price_df = price_df[~price_df['DistrictName'].str.contains('学区', na=False)]
filtered_price_count = len(price_df)
removed_count = initial_price_count - filtered_price_count
print(f"'DistrictName' に '学区' を含む行を {removed_count} 件削除しました。")

 4. 町丁目データの列名を変更
chochomoku_df_renamed = chochomoku_df.rename(columns={
    'X_CODE': 'longitude',
    'Y_CODE': 'latitude'
})

 必要な列のみを抽出し、コピーを作成してSettingWithCopyWarningを防ぐ
chochomoku_subset = chochomoku_df_renamed[['S_NAME', 'longitude', 'latitude']].copy()

 5. 部分一致によるマージ

 価格データのDistrictNameを文字列に変換
price_df['DistrictName'] = price_df['DistrictName'].astype(str)

 町丁目データのS_NAMEも文字列に変換
chochomoku_subset['S_NAME'] = chochomoku_subset['S_NAME'].astype(str)

def find_coordinates(district_name, chochomoku_df):
    """
    DistrictNameがS_NAMEに含まれるかをチェックし、対応する緯度経度を返す関数
    また、マッチの種類を示すフラグも返す
    """
    if not isinstance(district_name, str):
        return pd.Series({'longitude': None, 'latitude': None, 'match_type': 'invalid'})

     S_NAMEにDistrictNameが含まれる行をフィルタリング（部分一致）
    match = chochomoku_df[chochomoku_df['S_NAME'].str.contains(district_name, na=False)]

    if len(match) == 1:
         一意にマッチした場合
        return pd.Series({
            'longitude': match['longitude'].values[0],
            'latitude': match['latitude'].values[0],
            'match_type': 'single'
        })
    elif len(match) > 1:
         複数マッチした場合は最初のマッチを使用
        return pd.Series({
            'longitude': match['longitude'].values[0],
            'latitude': match['latitude'].values[0],
            'match_type': 'multiple'
        })
    else:
         マッチしない場合
        return pd.Series({
            'longitude': None,
            'latitude': None,
            'match_type': 'no_match'
        })

 価格データに対して部分一致で緯度経度を取得
match_results = price_df['DistrictName'].apply(lambda x: find_coordinates(x, chochomoku_subset))
price_df = pd.concat([price_df, match_results], axis=1)

 6. マッチング結果の集計
count_single = (price_df['match_type'] == 'single').sum()
count_multiple = (price_df['match_type'] == 'multiple').sum()
count_no_match = (price_df['match_type'] == 'no_match').sum()
count_invalid = (price_df['match_type'] == 'invalid').sum()

print(f"一意にマッチした件数: {count_single}")
print(f"複数マッチした件数: {count_multiple}")
print(f"マッチしなかった件数: {count_no_match}")
print(f"無効なDistrictNameの件数: {count_invalid}")

 7. マッチしなかった行に関する警告
if count_no_match > 0:
    missing_districts = price_df.loc[price_df['match_type'] == 'no_match', 'DistrictName'].unique()
    print(f"警告: 以下のDistrictNameに対応する町丁目データが見つかりませんでした: {missing_districts}")

if count_multiple > 0:
    multiple_districts = price_df.loc[price_df['match_type'] == 'multiple', 'DistrictName'].unique()
    print(f"警告: 以下のDistrictNameが複数のS_NAMEにマッチしました: {multiple_districts}")

if count_invalid > 0:
    invalid_districts = price_df.loc[price_df['match_type'] == 'invalid', 'DistrictName'].unique()
    print(f"警告: 以下のDistrictNameは無効です: {invalid_districts}")

 8. マージ結果の保存
try:
    price_df.to_csv(output_csv_path, index=False, encoding='utf-8-sig')
    print(f"マージされたデータを '{output_csv_path}' に正常に出力しました。")
except Exception as e:
    print(f"CSVへの出力中にエラーが発生しました: {e}")
    exit(1)
```
